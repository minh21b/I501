```python
import matplotlib.pyplot as plt
import time
```

```python
def loop_through(n):
    temp = []
    start_time = time.time()
    for i in range(n):
        temp.append(i)
    end_time = time.time()
    return end_time - start_time

def constant_time(n):
    start_time = time.time()
    result = n
    end_time = time.time()
    return end_time - start_time

def nested_loop(n):
    temp = []
    start_time = time.time()
    for i in range(n):
        for j in range(n):
            temp.append((i, j))
    end_time = time.time()
    return end_time - start_time
```

```python
# Values of n for plotting
n_values = list(range(1, 1001))

# Calculate time for loop_through function for each n
loop_times = [loop_through(n) for n in n_values]

# Calculate time for constant_time function for each n
constant_times = [constant_time(n) for n in n_values]

# Calculate time for nested_loop function for each n
nested_times = [nested_loop(n) for n in n_values]
```

```python
# Plotting
plt.figure(figsize=(10, 6))
plt.plot(n_values, loop_times, label='loop_through (O(n))')
plt.plot(n_values, constant_times, label='constant_time (O(1))')
plt.plot(n_values, nested_times, label='nested_loop (O(n^2))')
plt.xlabel('Input size (n)')
plt.ylabel('Time (s)')
plt.title('Comparison of Time Complexity')
plt.legend()
plt.grid(True)
plt.show()
```
